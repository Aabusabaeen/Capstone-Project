## Sheet 1-1

```python
Trading_data = pd.read_excel ('Data/Trading2020.xlsx', sheet_name='1-1')
Trading_data.to_csv("Trading2020_1-1.csv", sep=",") 
# as seen below, this method has only used the data in the first sheet and ignored others
```

```python
Trading_data
```

```python
Trading_data.drop([0,1,2], axis=0, inplace=True)
```

```python
Trading_data.drop('R', axis=1, inplace=True)
```

```python
Trading_data.drop([4], axis=0, inplace=True)
```

```python
Trading_data.rename(columns={'Unnamed: 0': 'Year', 'Unnamed: 1': 'Imports Weight(1000 Tons)',
 'Unnamed: 2' : 'Imports Vaule(Millions SAR)', 'Unnamed: 3' : 'Exports Weight(1000 Tons)', 'Unnamed: 4' : 'Exports Vaule(Millions SAR)'}, inplace= True)
```

```python
Trading_data.drop([3,5], axis=0, inplace=True)
```

```python
Cleaned_Trading20 = Trading_data.copy(deep= True)
```

```python
# same as above
Cleaned_Trading20 = pd.DataFrame(Trading_data)
```

```python
Cleaned_Trading20['Difference in Weights']= Cleaned_Trading20['Exports Weight(1000 Tons)'] - Cleaned_Trading20['Imports Weight(1000 Tons)']
Cleaned_Trading20
```

```python
Cleaned_Trading20['Difference in Value']= Cleaned_Trading20['Exports Vaule(Millions SAR)'] - Cleaned_Trading20['Imports Vaule(Millions SAR)']
Cleaned_Trading20
```

### Difference in Weights of Tradings over the years
