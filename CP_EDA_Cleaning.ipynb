Imports by Transoprtation Type

```python
Imports_byTransoprtType = pd.read_excel ('Data/Trading2020.xlsx', sheet_name='1-5')
Imports_byTransoprtType.to_csv("Trading2020_1-5.csv", sep=",") 
```

```python
Imports_byTransoprtType
```

```python
Imports_byTransoprtType.rename(columns={'الواردات حسب وسيلة النقل' :  'Weight (1000 Tons) 2018',
'Unnamed: 2' : '2018 Value (millions in SAR)', 'Unnamed: 3' : '2019 Weight (1000 Tons)', 'Unnamed: 4' : '2019 Value (millions in SAR)', 
'Unnamed: 5' : '2020 Weight (1000 Tons)', 'Unnamed: 6' : '2020 Value (millions in SAR)', 'Unnamed: 7' : 'Importing Type'}, inplace=True)
```

```python
Imports_byTransoprtType.drop([0,1,2,3,4], axis=0, inplace=True)
Imports_byTransoprtType.drop(['Unnamed: 0', 'R'], axis=1, inplace=True)
Imports_byTransoprtType.set_index([pd.Index()], inplace=True)
```

```python
Imports_byTransoprtType.drop([8], axis=0, inplace=True)
```

```python
Imports_byTransoprtType
```

```python
Imports_byTransoprtType[["Weight (1000 Tons) 2018", "2019 Weight (1000 Tons)", "2020 Weight (1000 Tons)"]].sum()
```

Source to navigate the first cell value in the columns:
https://www.stackvidhya.com/get-value-of-cell-from-a-pandas-dataframe/

```python
plt.style.use('seaborn-darkgrid') # to have a background that makes the chart more visible
x = Imports_byTransoprtType[["Weight (1000 Tons) 2018", "2019 Weight (1000 Tons)", "2020 Weight (1000 Tons)"]].values[0]
labels1= ['2018', '2019', '2020']
plt.figure(figsize = [8, 8])
plt.pie(x, labels = labels1, counterclock= False, autopct= lambda p:f'{p:.2f}%')
plt.title('Imports by Sea from 2018-2020', fontsize= 17)
plt.legend(Imports_byTransoprtType[["Weight (1000 Tons) 2018", "2019 Weight (1000 Tons)", "2020 Weight (1000 Tons)"]].values[0], 
loc='upper left', fontsize= 10)
plt.show()
```

```python
x = ['2018', '2019', '2020']
sea = Imports_byTransoprtType[["Weight (1000 Tons) 2018", "2019 Weight (1000 Tons)", "2020 Weight (1000 Tons)"]].values[0]
land = Imports_byTransoprtType[["Weight (1000 Tons) 2018", "2019 Weight (1000 Tons)", "2020 Weight (1000 Tons)"]].values[1]
air = Imports_byTransoprtType[["Weight (1000 Tons) 2018", "2019 Weight (1000 Tons)", "2020 Weight (1000 Tons)"]].values[2]
sea_land = list(np.add(sea,land))
plt.figure(figsize = [7,6])
plt.bar(x, sea, 0.5, label= 'Sea')
plt.bar(x, land, 0.4, bottom= sea, label= 'Land')
plt.bar(x,air, 0.3, bottom= sea_land, label= 'Air')
plt.title('Weight of Imports by Transportation Types', fontsize= 17)
plt.xlabel('Year',fontsize= 15)
plt.xticks(fontsize= 13)
plt.yticks(fontsize= 13)
plt.ylabel('Weights in 1000 Tons', fontsize= 15)
plt.legend(loc= 'upper left', fontsize= 12)
plt.show()
```
